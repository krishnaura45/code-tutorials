## <b><span style='color:orange'>STATISTICAL ANALYSIS</span></b> 

**1. Loading Dataset**

```python
import pandas as pd
data=pd.read_csv('/kaggle/input/aiml-dataworld/cancerdata.csv')
display(data)
```

**2. Calculating Mean Median and Mode of an attribute**

```python
import numpy as np
import statistics as st
attr='radius_mean'
mean=data[attr].mean(axis=0)
median=np.median(data[attr])
mode=st.mode(data[attr])
print('Attribute : '+attr)
print('Mean of above attribute')
display(mean)
print('Median of above attribute')
display(median)
print('Mode of above attribute')
display(mode)
```

**3. Calculating Range and Midrange of an attribute**

```python
m=data[attr]
min=np.min(m)
max=np.max(m)
r=max-min
midrange=(max+min)/2
print('Attribute : '+attr)
print('Range of attribute')
print(r)
print('Midrange of attribute')
print(midrange)
```

**4. Calculating Variance and Standard Deviation of an attribute**

```python
var=st.variance(data[attr])
stddev=np.std(data[attr])
print('Attribute : '+attr)
print('variance of attribute')
print(var)
print('standard deviation of attribute')
print(stddev)
```

**5. Calculating Covariance between two attributes**

```python
attr1='radius_mean'
attr2='perimeter_mean'
covariance=np.cov(data[attr1],data[attr2])
print('Attribute 1 : '+attr1)
print('Attribute 2 : '+attr2)
print('Covariance of two attributes')
print(covariance)
```

**6. Calculating Correlation between two attributes**

```python
correlation=np.corrcoef(data[attr1],data[attr2])
print('Attribute 1 : '+attr1)
print('Attribute 2 : '+attr2)
print('Correlation of two attributes')
print(correlation)
```

**7. Calculating 5 Number summary of an attribute**

```python
from numpy import percentile
quartiles=percentile(data[attr],[25,50,75])

print("The five number summary of ",attr,' is :')
print ( 'Min: %.3f' % min)
print( 'Q1: %.3f' % quartiles[0] )
print ( 'Median: %.3f' % quartiles[1] )
print( 'Q3: %.3f' % quartiles[2] )
print ( 'Max: %.3f' % max)
```

**8. Identifying attribute types in dataset**

```python
data.select_dtypes(include=np.number)
data.info()
data.head(10)
# id is numerical(discrete) attribute
# diagnosis is nominal and binary attribute
# rest all are continuous attributes
```

## <b><span style='color:orange'>SAMPLING</span></b> 

```python
# loading a different dataset from the same repository
data=pd.read_csv('/kaggle/input/aiml-dataworld/cancerdata.csv')
display(data)
```

**1. Creation of a sampled data set using random sampling without replacement**

```python
# Selecting columns I am interested in
columns= ['id','diagnosis','radius_mean','texture_mean','area_mean','perimeter_mean','texture_worst','perimeter_worst','area_worst']
data = data.loc[:, columns]
s=data.sample(n = 460, replace = False, random_state=100)
print("Sample Dataset")
display(s)
```

**2. Creation of a sampled data set using random sampling with replacement**

```python
sr=data.sample(n = 5400, replace = True, random_state=100)
print("Sample")
display(sr)
```

**3. Creation of sampled datasets using cluster sampling**

```python
# Dividing the dataset into clusters based on 'diagnosis'
clusters = data.groupby('diagnosis')

# selecting the 'M' cluster
selected_clusters = clusters.get_group('M')

# Display the selected cluster
print("Selected Cluster (M):")
print(selected_clusters)

# The resulting selected_clusters DataFrame is the required sample
```

In practice, you might want to randomly select clusters rather than manually choosing one.

```python
# Divide the dataset into clusters based on 'diagnosis'
clusters = data.groupby('diagnosis')

# Convert the cluster keys to a list
cluster_keys = list(clusters.groups.keys())

# Randomly select clusters (e.g., select 1 cluster)
selected_cluster_labels = np.random.choice(cluster_keys, size=1, replace=False)

# Create the sampled dataset from the selected clusters
sampled_df = pd.concat([clusters.get_group(label) for label in selected_cluster_labels])

# Display the sampled dataset
print("Sampled Dataset:")
print(sampled_df)
```

**4. Creation of a sampled data set using stratified sampling**

```python
# Using pandas groupby, separating the values into groups based on their diagnosis i.e M, B 
# and randomly sample 5 values from each diagnosis group using the sample function
data.groupby('diagnosis', group_keys=False).apply(lambda x: x.sample(5))
```

```python
# Using pandas groupby, separate the data values into groups based on their diagnosis i.e M, B 
# and randomly sample from each group based on population proportion. 
# (Assuming the total sample size to be 60%(0.6) of the population)
data.groupby('diagnosis', group_keys=False).apply(lambda x: x.sample(frac=0.6))
```

## <b><span style='color:orange'>ENCODING</span></b> 

```python
# dataset no. 2
data=pd.read_csv('/kaggle/input/aiml-dataworld/pokemon.csv')
display(data)
```

***Discretization***

```python
from sklearn.preprocessing import KBinsDiscretizer

# Selecting the dataframe part having numerical attributes
df1= data.iloc[:,4:11]

# Performing K-means Discretization 
discretizer = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')
data_trans = discretizer.fit_transform(df1)

display(pd.DataFrame(data_trans))
```

```python
# histogram of the transformed data
import matplotlib.pyplot as plt
plt.hist(data_trans, bins=10)
plt.show()
```

***Label Encoding***

```python
def encode_target(df, target_column):
    df_mod = df.copy()
    targets = np.unique(df_mod[target_column]) 
    
    # nominal to numeric conversion
    map_to_int = {name: n for n, name in enumerate(targets)}
    df_mod["Target"] = df_mod[target_column].replace(map_to_int)  

    return (df_mod, targets)

encode_target(data,'Type 1')
```

***One Hot Encoding***

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
  
# Converting attribute 'Generation' to categorical type
data['Generation']=data['Generation'].astype('category')
  
  
#Assigning numerical values and storing it in another columns
data['GEN_new']=data['Generation'].cat.codes
  
  
#Create an instance of One-hot-encoder
enc=OneHotEncoder()
  
#Passing encoded column of Generation
'''
NOTE: we have converted the enc.fit_transform() method to array because the fit_transform method 
of OneHotEncoder returns SpiPy sparse matrix this enables us to save space when we 
have huge  number of categorical variables
'''
New_df=enc.fit_transform(data[['GEN_new']]).toarray()
ohe_data=pd.DataFrame(New_df,columns=['Gen_1','Gen_2','Gen_3','Gen_4','Gen_5','Gen_6'])
  
display(ohe_data)
```

***Dummy Encoding***

```python
de_data = ohe_data.iloc[:,1:6]
display(de_data)

# Dummy encoding is similar to one hot encoding but it creates n-1 columns after encoding,
# where n is the number of categories, 
# while in one hot encoding we have n columns.
```

## <b><span style='color:orange'>NORMALIZATION</span></b> 

```python
# create another dataset 
df2 = pd.DataFrame([
                   [180000, 110, 18.9, 1400], 
                   [360000, 905, 23.4, 1800], 
                   [230000, 230, 14.0, 1300], 
                   [60000, 450, 13.5, 1500]], 
    
                   columns=['Col A', 'Col B',
                            'Col C', 'Col D'])
  
# view data
display(df2)
```

**Min-max normalization**

```python
# copy the data
df_min_max_scaled = df2.copy()
  
# apply normalization techniques
for col_x in df_min_max_scaled.columns:
    df_min_max_scaled[col_x] = (df_min_max_scaled[col_x] - df_min_max_scaled[col_x].min()) / (df_min_max_scaled[col_x].max() - df_min_max_scaled[col_x].min())    
  
# view normalized data
print(df_min_max_scaled)
```

**Decimal Scaling**

```python
# Applying decimal scaling on 'Col A' attribute values
max=df2['Col A'].max()
digit=len(str(round(max)))
div=10**2
print(df2['Col A']/div)
```

**Z Score Normalization**

```python
# Applying Z score scaling on 'Col B' attribute values
from sklearn.preprocessing import StandardScaler

z_score = StandardScaler()
data = z_score.fit(df2[['Col B']])
scaling = data.transform(df2[['Col B']])
scaling
```

## <b><span style='color:orange'>HANDLING MISSING VALUES</span></b> 

**Creating dataset with missing values**

```python
dict = {'First Score':[100, 90, np.nan, 95,87],
		'Second Score': [30, 45, np.nan, np.nan,67],
		'Third Score':[np.nan, 40, np.nan, 98,74]}

# creating a dataframe 
df = pd.DataFrame(dict)
display(df)
```

**Checking for missing values using isnull()**

```python
df.isnull()    # isnull() function returns dataframe of Boolean values which are True for NaN values
```

### Ways to handle missing values
#### 1) **Ignoring the tuple**

**Dropping rows with atleast one nan value**

```python
# using dropna() function 
df.dropna()
```

**Dropping rows if all values in that row are missing**

```python
df.dropna(how = 'all')
```

#### 2) **Filling the missing values using global constant**

```python
# will replace  Nan value in dataframe with value 0 
df.replace(to_replace = np.nan, value = 0)
```

#### 3) **Filling the missing values manually**

```python
ndf=df
# filling null values using fillna()
ndf.fillna("Absent", inplace = True)
display(ndf)
```

#### 4) **Mean, Median and Mode Imputation (filling missing values with the measures of Central tendency)**

```python
# Ensure columns are the correct data type
df['First Score'] = pd.to_numeric(df['First Score'], errors='coerce')
df['Second Score'] = pd.to_numeric(df['Second Score'], errors='coerce')

# Replace the null values of column 1 with the mean
df['First Score'] = df['First Score'].fillna(df['First Score'].mean())

# Replace column 2 null values with the median
df['Second Score'] = df['Second Score'].fillna(df['Second Score'].median())
print(df)
```

### Making the data smooth by Binning

```python
import numpy as np
from sklearn.datasets import load_iris

# load iris data set
dataset = load_iris()
a = dataset.data
b = np.zeros(150)

# take 1st column among 4 columns of data set
for i in range(150):
    b[i] = a[i, 1]

b = np.sort(b)  # sort the array

# create bins
bin1 = np.zeros((30, 5))
bin2 = np.zeros((30, 5))
bin3 = np.zeros((30, 5))

# Bin mean
for i in range(0, 150, 5):
    k = int(i / 5)
    mean = np.mean(b[i:i+5])
    bin1[k, :] = mean
print("Bin Mean: \n", bin1)

# Bin boundaries
for i in range(0, 150, 5):
    k = int(i / 5)
    for j in range(5):
        if (b[i + j] - b[i]) < (b[i + 4] - b[i + j]):
            bin2[k, j] = b[i]
        else:
            bin2[k, j] = b[i + 4]
print("\nBin Boundaries: \n", bin2)

# Bin median
for i in range(0, 150, 5):
    k = int(i / 5)
    median = np.median(b[i:i+5])
    bin3[k, :] = median
print("\nBin Median: \n", bin3)
```
