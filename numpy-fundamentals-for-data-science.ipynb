## <b><span style='color:blue'>Features of Numpy</span></b>

Numpy's most useful feature is **n dimensional array object**, also known as ***ndarray***

```python
a=np.array([1,2,3])
a
```

```python
a[2]
```

Three main benefits of using a *numpy array* over a *python list* are-
- less memory
- fast
- convinient

### Less memory

```python
import time 
import sys

l=range(1000)
print(sys.getsizeof(6)*len(l))   # number of bytes taken by python list

array=np.arange(1000)
print(array.size * array.itemsize)   # no of bytes taken by nd array
```

1 'int64' element takes 8 bytes of memory in numpy array whereas 28 bytes of memory in python list

### Fast 

```python
# We will add two list of numbers using python list and numpy array and compare them
SIZE=1000000
l1=range(SIZE)
l2=range(SIZE)

a1=np.arange(SIZE)
a2=np.arange(SIZE)

#python list
start=time.time()
result=[(x+y) for x,y in zip(l1,l2)]
print("python list took:",(time.time()-start)*1000)

#numpy array
start2=time.time()
result=a1+a2
print("numpy took:",(time.time()-start2)*1000)
```

### Convinient

```python
a1=np.array([1,2,3])
a2=np.array([4,5,6])
print(a1+a2)
print(a1-a2)
print(a1/a2)
print(a1*a2)
```

```python
a[-2:-1]   # reverse indexing
```

```python
arr[0:2,2]
```

```python
arr[-1]  # last row
```

## <b><span style='color:blue'>Iterating through arrays</span></b>

```python
arr
```

```python
# print each row independently
for row in arr:
    print(row)
```

```python
# flatten the list and print each cell
for cell in arr.flat:
    print(cell)
```

```python
# same operation as above
for cell in arr.ravel():
    print(cell)
```

```python
# same operation as above
for cell in arr.flatten():
    print(cell)
```

**Iteration using nditer**

```python
# in C-order (row/cell wise traversal)
for x in np.nditer(arr,order='C'):
    print(x)
```

```python
# in Fortran order (column wise traversal)
for x in np.nditer(arr,order='F'):
    print(x)
```

Supply flags on for loop

```python
# print entire column on each iteration
for x in np.nditer(arr,order='F',flags=['external_loop']):
    print(x)
```

```python
for x in np.nditer(arr,op_flags=['readwrite']):
    x[...]=x*x
arr
```

```python
# iterate through two arrays arr and brr simultaneously
# only possible when dimns of both the arrays are compatible
# that is, either dimensions (shapes) equal or one of them is 1 
brr=np.arange(3,15,4).reshape(3,1)
brr
```

```python
for x,y in np.nditer([arr,brr]):
    print(x,y)
```

## <b><span style='color:blue'>Stack/Split multiple arrays</span></b>

```python
a=np.arange(6).reshape(3,2)
b=np.arange(6,12).reshape(3,2)
```

```python
# vertical stacking
np.vstack((a,b))
```

```python
# horizontal stacking
np.hstack((a,b))
```

```python
#horizontal split
a=np.arange(12).reshape(3,4)
a
```

```python
np.hsplit(a,2)
```

```python
# vertical split
np.vsplit(a,3)
```
