## 5. Sales Data

## 5.1. Merging various CSV files into single CSV dataframe

```python
import numpy as np
import os

# merging twelve csv files of sales data
sales_df = pd.concat(
    map(pd.read_csv, ['/kaggle/input/aiml-dataworld/Company_monthly_sales/jan.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/feb.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/march.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/april.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/may.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/june.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/july.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/aug.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/sep.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/oct.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/nov.csv','/kaggle/input/aiml-dataworld/Company_monthly_sales/dec.csv']), ignore_index=True)
display(sales_df)
```

## 5.2. Dropping all NaN values from the dataframe

```python
# Finding out the dataset containing NaN values
nan_df = sales_df[sales_df.isna().any(axis=1)]
display(nan_df)
print("\n")

# Finding the new dataset
sales_df = sales_df.dropna()
display(sales_df)
```

## 5.3. Adding new column in data frame

```python
# Creating a column of month
sales_df['Month'] = sales_df['Order Date'].str[0:2]
display(sales_df)
```

```python
sales_df['Purchase Address'].str[:]
```

```python
# Creating a column of city
# Splitting the 'Purchase Address' column by comma and extracting the city
sales_df['City'] = sales_df['Purchase Address'].str.split(',').str.get(-2)
display(sales_df)
```

## 5.4. Finding the best month for sales

```python
sales_df['Quantity Ordered'].sample(5)
```

```python
sales_df['Price Each'].sample(5)
```

```python
# Convert 'Quantity Ordered' and 'Price Each' columns to numeric, ignoring errors
sales_df['Quantity Ordered'] = pd.to_numeric(sales_df['Quantity Ordered'], errors='coerce')
sales_df['Price Each'] = pd.to_numeric(sales_df['Price Each'], errors='coerce')

# Drop rows where either 'Quantity Ordered' or 'Price Each' is NaN (indicating parsing error)
sales_df.dropna(subset=['Quantity Ordered', 'Price Each'], inplace=True)

# Calculate sales for each row
sales_df['Sales'] = sales_df['Quantity Ordered'] * sales_df['Price Each']

# Group by month and sum up the sales
monthly_sales = sales_df.groupby('Month')['Sales'].sum()

# Find the month with the highest sales
best_month = monthly_sales.idxmax()  # Returns the index of the maximum value, which corresponds to the month
bm=int(best_month)
print("The best month for sales is",best_month,"with sales",monthly_sales.iloc[bm -1],"$")
```

```python
import matplotlib.pyplot as plt     # data visualisation module
months = range(1,13)
plt.bar(months,monthly_sales)
plt.xticks(months)
plt.ylabel('Sales in US $')
plt.xlabel('Month number')
plt.show()
```

## 5.5. Finding the city that sold the most product

```python
# Group by city and sum up the quantity ordered
city_sales = sales_df.groupby('City')['Quantity Ordered'].sum()
```

```python
city_sales
```

```python
# Find the city with the highest total quantity of products sold
best_selling_city = city_sales.idxmax()  # Returns the index of the maximum value, which corresponds to the city
print("The city that sold the most products is:", best_selling_city)
```

## 5.6. Best Time to display advertisements to maximize likelihood of customer's buying product

```python
# Convert 'Order Date' column to datetime format
sales_df['Order Date'] = pd.to_datetime(sales_df['Order Date'])

# Extract hour from 'Order Date' and create a new column
sales_df['Order Hour'] = sales_df['Order Date'].dt.hour

# Group by hour and count the number of orders
hourly_orders = sales_df.groupby('Order Hour').size()

# Find the hour(s) with the highest number of orders
best_advertisement_hours = hourly_orders[hourly_orders == hourly_orders.max()].index.tolist()
print("The best time(s) to display advertisements to maximize the likelihood of customer's buying products:")
for hour in best_advertisement_hours:
    print(f"{hour}:00 to {hour + 1}:00")
```

## 5.7. Products most often sold together

```python
df1 = sales_df[sales_df['Order ID'].duplicated(keep=False)]
df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df1[['Order ID', 'Grouped']].drop_duplicates()

from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    display(key, value)
```

## 5.8. Column containing the highest number of row-wise maximum values

```python
# Convert columns to numeric type
sales_df = sales_df.apply(pd.to_numeric, errors='coerce')

max_column_counts = {}

for index, row in sales_df.iterrows():
    max_value_index = row.values.argmax()  # Find index of maximum value in the row
    if max_value_index not in max_column_counts:
        max_column_counts[max_value_index] = 1
    else:
        max_column_counts[max_value_index] += 1

# Find the column index with the highest count of occurrences
column_with_highest_count = max(max_column_counts, key=max_column_counts.get)

# Get the column name corresponding to the index
column_name = sales_df.columns[column_with_highest_count]
print(column_name)
```
