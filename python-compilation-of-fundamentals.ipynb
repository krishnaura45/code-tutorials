## Functions

A Function is a **block** of code that performs a **specific task**. For example, a *dish washer* takes dirty dishes as *input* and gives clean dishes as *output*.

```python
# Suppose you are given with two lists of numbers and you need to find total of each of these list
tom_exp_list=[2100,3400,3500]
joe_exp_list=[200,500,700]
def sum_list(list):
    total=0
    for i in list:
        total+=i;
    return total

print("Tom's total expense:",sum_list(tom_exp_list))
print("Joe's total expense:",sum_list(joe_exp_list))
```

Named vs Positioned Arguments

```python
# Sum of two numbers
def sum(a,b):
    print("a:",a)
    print("b:",b)
    
    total=a+b
    print("total inside function:",total)
    return total

n=sum(5,6)  #positioned
print("Total outside function:",n)
```

```python
def sum(a,b):
    print("a:",a)
    print("b:",b)
    
    total=a+b
    print("total inside function:",total)
    return total

n=sum(b=5,a=6)  #named
print("Total outside function:",n)
```

Default argument

```python
def sum(a,b=2):    # b is default argument
    print("a:",a)
    print("b:",b)
    
    total=a+b
    return total

n=sum(7)          # affect only when not passed
print("Total:",n)
```

```python
def sum(a,b=2):    # b is default argument
    print("a:",a)
    print("b:",b)
    
    total=a+b
    return total

n=sum(7,3)         # work normally if passed
print("Total:",n)
```

Local vs Global variables

```python
total=0    # global variable
def sum(a,b):
    print("a:",a)
    print("b:",b)
    total=a+b   # local variable
    print("total inside function:",total)
    return total

n=sum(b=5,a=6)
print("Total outside function:",total)
```

## Read, write files

**File Open Modes**:
- r: opens file for reading only. Throws error if file does not exist.
- w: opens file for writing only. If file doesn't exist then it will create new one. If it exists, it will overwrite it.
- r+: opens file for both reading and writing. If file doesn't exist, it will not create anything.
- w+: opens file for reading and writing. If file doesn't exist then it will create new one. If exists then it will overwrite it.
- a: opens a file in append mode. Whatever you write to file will get appended and original file/content will not be overwritten.


**Creating a new file and writing something into that file**

```python
f=open("/kaggle/working/funny.txt",'w')
f.write("I love Python")
f.close()
```

```python
# overwriting to the same file
f=open("/kaggle/working/funny.txt",'w')
f.write("I like Python")
f.close()
```

```python
# appending to the same file
f=open("/kaggle/working/funny.txt",'a')
f.write("I like C++")
f.close()
```

```python
# appending to the same file but in different line using '\n'
f=open("/kaggle/working/funny.txt",'a')
f.write("\nI like Java")
f.close()
```

**Read a file**

```python
# reading the file as a whole
f=open("/kaggle/working/funny.txt",'r')
print(f.read())
f.close()
```

```python
# reading the file line by line
f=open("/kaggle/working/funny.txt",'r')
for line in f:
    print(line)
f.close()
```

```python
# Reading and writing simultaneously but in different files
# That is, reading one file and writing into another at the same time

# Let's concatenate word count at the start of each line
f=open("/kaggle/working/funny.txt",'r')
f1=open("/kaggle/working/funnywc.txt",'w')
for line in f:
    tokens=line.split(' ')   # splitting by space --> gives a list of words in line
    print(str(tokens))
    print(len(tokens))
    f1.write('wc: '+str(len(tokens))+' -> '+line)
f.close()
f1.close()
```

```python
# reading the file as a whole
f1=open("/kaggle/working/funnywc.txt",'r')
print(f1.read())
f1.close()
```

## Classes and Objects

Defining class

```python
class Human:
    # Initializer
    def __init__(self, n, o):
        self.name = n
        self.occupation = o
    
    # Method 1
    def do_work(self):
        if self.occupation == "tennis player":
            print(self.name, "plays tennis")
        elif self.occupation == "actor":
            print(self.name, "shoots film")
    
    # Method 2
    def speaks(self):
        print(self.name, "says how are you?")
```

Objects

```python
amir = Human("aamir khan","actor")
amir.do_work()
amir.speaks()
```

```python
sania = Human("sania mirza","tennis player")
sania.do_work()
sania.speaks()
```

## Inheritance

```python
# Parent class
class Vehicle:
    def general_usage(self):
        print("general use: transporation")

# Sub classes or Inherited classes
class Car(Vehicle):
    def __init__(self):
        print("I'm car")
        self.wheels = 4
        self.has_roof = True

    def specific_usage(self):
        self.general_usage()
        print("specific use: commute to work, vacation with family")

class MotorCycle(Vehicle):
    def __init__(self):
        print("I'm motor cycle")
        self.wheels = 2
        self.has_roof = False

    def specific_usage(self):
        self.general_usage()
        print("specific use: road trip, racing")

# Istances of sub classes
c = Car()
m = MotorCycle()
```

```python
print(issubclass(Car,Vehicle))
```

```python
print(issubclass(MotorCycle,Vehicle))
```

```python
print(issubclass(Car,MotorCycle))
```

## Iterators

```python
# Custom Iterator
class RemoteControl():
    def __init__(self):
        self.channels = ["HBO","cnn","abc","espn"]
        self.index = -1

    def __iter__(self):
        return self

    def __next__(self):
        self.index += 1
        if self.index == len(self.channels):
            raise StopIteration

        return self.channels[self.index]

r = RemoteControl()
itr=iter(r)
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
```
