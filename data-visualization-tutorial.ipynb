## <b><span style='color:green'>Getting Started</span></b>

```python
# supportive library
import pandas as pd
```

Matplotlib and Seaborn are two popular Python libraries used for data visualization.

**Matplotlib**: A comprehensive library for creating static, animated, and interactive visualizations in Python.
- It provides a wide variety of plotting functions and customization options to create plots such as line plots, scatter plots, bar plots, histograms, pie charts, and more.
- Matplotlib is highly customizable, allowing users to control every aspect of their plots, including colors, labels, annotations, and axes.
- It is the foundation for many other Python plotting libraries and is widely used in the scientific and data analysis communities.


```python
from matplotlib import pyplot as plt
```

**Seaborn**: A statistical data visualization library built on top of Matplotlib.
- It provides a high-level interface for creating attractive and informative statistical graphics.
- Seaborn simplifies the process of creating complex plots by providing functions for common visualization tasks such as scatter plots, box plots, violin plots, pair plots, and heatmaps.
- Seaborn offers built-in support for visualizing categorical data, fitting regression models, and displaying distributions.
- It also comes with built-in themes and color palettes to quickly change the overall look and feel of plots.


```python
import seaborn as sns
```

## <b><span style='color:green'>Line Chart</span></b>

```python
df_sales = pd.read_excel("/kaggle/input/aiml-dataworld/appliance_sales.xlsx")
df_sales
```

```python
df_sales.columns[1:]
```

<a id="2.1"></a>
### <b><span style='color:green'>2.1 Using Matplotlib</span></b>

```python
plt.figure(figsize=(12,6))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color="blue", label="Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"], color="orange", label="Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"], color="green", label="Washing Machine")

plt.title("Product Sales")
plt.xlabel("Quarter")
plt.ylabel("Million $")

plt.legend()
plt.show()
```

```python
import numpy as np
df_sales['Quarter'].values
```

```python
quarters = df_sales['Quarter'].values
fridge = df_sales['Fridge'].values
dishwasher = df_sales['Dishwasher'].values
washing_machine = df_sales['Washing Machine'].values

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot lines
ax.plot(quarters, fridge, marker='o', label='Fridge')
ax.plot(quarters, dishwasher, marker='o', label='Dishwasher')
ax.plot(quarters, washing_machine, marker='o', label='Washing Machine')

# Add labels and title
ax.set_xlabel('Quarter')
ax.set_ylabel('Sales')
ax.set_title('Product Sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

Adding **linestyle** ,**color**, **annotation** and **axis limit**

```python
# Create a figure and axis object
fig, ax = plt.subplots()

# Plot lines with custom styles and colors
ax.plot(quarters, fridge, marker='o', linestyle='-', color='blue', label='Fridge')
ax.plot(quarters, dishwasher, marker='o', linestyle='--', color='green', label='Dishwasher')
ax.plot(quarters, washing_machine, marker='o', linestyle='-.', color='red', label='Washing Machine')

# Add labels and title
ax.set_xlabel('Quarter')
ax.set_ylabel('Sales')
ax.set_title('Appliance Sales')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Add grid lines
plt.grid(True)

# Add annotations
for i, txt in enumerate(fridge):
    ax.annotate(txt, (quarters[i], fridge[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Set axis limits
ax.set_ylim(0, 20)

# Display the plot
plt.tight_layout()
plt.show()
```

### <b><span style='color:green'>2.2 Using Seaborn</span></b>

```python
# Melt the DataFrame
df_melted = df_sales.melt(id_vars='Quarter', var_name='Product', value_name='Sales')

# Plot using Seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_melted, x='Quarter', y='Sales', hue='Product')
plt.title('Product Sales')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

In above code:

- Firstly, I melted the DataFrame using **melt()** function to convert it to long format, which is suitable for Seaborn's lineplot.
- Finally, I plotted the line chart using Seaborn's **lineplot()** function, specifying the x-axis, y-axis, and hue (for differentiating lines by appliance), along with other optional parameters like marker style, title, labels, rotation of x-axis labels, and grid lines.

Seaborn makes it straightforward to create informative and visually appealing line charts with just a few lines of code.

***Customizing*** - changing **color palette**, adding **error bars**, **smoothing lines** and **marking**

```python
# Melt the DataFrame to have 'Quarter' as the x-axis and 'value' as the y-axis
df_melted = df_sales.melt(id_vars='Quarter', var_name='Appliance', value_name='Sales')

# Set seaborn style and color palette
sns.set(style="darkgrid")
sns.set_palette("husl")

# Create the line plot with error bars and smoothing
plt.figure(figsize=(8, 6))
sns.lineplot(data=df_melted, x='Quarter', y='Sales', hue='Appliance', marker='^',markersize='8', errorbar='sd', estimator='mean')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Set title and labels
plt.title('Appliance Sales')
plt.xlabel('Quarter')
plt.ylabel('Sales')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create the facet grid
g = sns.FacetGrid(df_melted, col="Appliance", col_wrap=2, height=5)
g.map(sns.lineplot, "Quarter", "Sales", marker='o', linestyle='--')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Set title and labels
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Appliance Sales Over Quarters')

# Show plot
plt.tight_layout()
plt.show()
```

In this example:

- We use **sns.FacetGrid()** to create a facet grid with col_wrap=2, meaning that it will have 2 columns and the number of rows will be determined based on the number of appliances. We also set the height of each subplot to 5 units.
- **g.map()** is used to apply the **sns.lineplot()** function to each facet (subplot) of the grid

## <b><span style='color:green'>Bar Chart</span></b>

<a id="4.1"></a>
### <b><span style='color:green'>4.1 Using Matplotlib</span></b>

#### Simplest way

```python
df_sales.plot(kind="bar", x="Quarter")

plt.xticks(rotation=45)
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2 = df_sales.set_index("Quarter")
df_sales_2
```

```python
df_sales_2.plot(kind="bar")
plt.xticks(rotation=45)
plt.show()
```

```python
# Horizontal Bar Chart
df_sales_2.plot(kind="barh")
plt.show()
```

#### Best way

```python
# Getting the data columns individually
quarters = df_sales['Quarter']
fridge_sales = df_sales['Fridge']
dishwasher_sales = df_sales['Dishwasher']
washing_machine_sales = df_sales['Washing Machine']

# Setting the width of the bars
bar_width = 0.25

# Setting the position of the bars on the x-axis
r1 = range(len(quarters))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Preparig the bar chart
plt.figure(figsize=(10, 6))
plt.bar(r1, fridge_sales, color='skyblue', width=bar_width, edgecolor='grey', label='Fridge')
plt.bar(r2, dishwasher_sales, color='lightgreen', width=bar_width, edgecolor='grey', label='Dishwasher')
plt.bar(r3, washing_machine_sales, color='salmon', width=bar_width, edgecolor='grey', label='Washing Machine')

# Add xticks on the middle of the group bars
plt.xlabel('Quarters', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(quarters))], quarters, rotation=45)

plt.ylabel('Sales', fontweight='bold')
plt.title('Appliance Sales', fontweight='bold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Setting the height of the bars
bar_height = 0.25

# Setting the position of the bars on the y-axis
r1 = range(len(quarters))
r2 = [y + bar_height for y in r1]
r3 = [y + bar_height for y in r2]

# Preparing the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(r1, fridge_sales, color='skyblue', height=bar_height, edgecolor='grey', label='Fridge')
plt.barh(r2, dishwasher_sales, color='lightgreen', height=bar_height, edgecolor='grey', label='Dishwasher')
plt.barh(r3, washing_machine_sales, color='salmon', height=bar_height, edgecolor='grey', label='Washing Machine')

# Adding yticks on the middle of the group bars
plt.ylabel('Quarters', fontweight='bold')
plt.yticks([r + bar_height for r in range(len(quarters))], quarters)

plt.xlabel('Sales', fontweight='bold')
plt.title('Appliance Sales by Quarter', fontweight='bold')
plt.legend()

plt.tight_layout()
plt.show()
```

## <b><span style='color:green'>Histogram</span></b>

```python
df_score = pd.read_csv("/kaggle/input/aiml-dataworld/student_exam_performance.csv")
df_score.head()
```

<a id="5.1"></a>
### <b><span style='color:green'>5.1 Using Matplotlib</span></b>

#### Simple start

```python
# histogram using matplotlib
plt.hist(df_score["Exam_Score"], density=True)
plt.show()
```

#### Better way

```python
exam_scores = df_score['Exam_Score']

plt.figure(figsize=(10, 6))
plt.hist(exam_scores, bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Best way

```python
plt.figure(figsize=(10, 6))
plt.hist(exam_scores, bins=10, color='skyblue', edgecolor='black', alpha=0.7)  # Set alpha for transparency

# Add vertical line for average score
average_score = exam_scores.mean()
plt.axvline(average_score, color='red', linestyle='dashed', linewidth=1, label=f'Average Score: {average_score:.2f}')

plt.title('Distribution of Exam Scores', fontsize=16)
plt.xlabel('Exam Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## <b><span style='color:green'>Scatter Plot</span></b>

```python
df = pd.read_csv("/kaggle/input/aiml-dataworld/property_valuation.csv")
df
```

<a id="6.1"></a>
### <b><span style='color:green'>6.1 Using Matplotlib</span></b>

```python
area = df['area_square_ft']
price = df['price']

plt.figure(figsize=(10, 6))
plt.scatter(area, price, color='skyblue', edgecolor='black')

plt.title('Area vs. Price')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')

plt.grid(True)

plt.tight_layout()
plt.show()
```

<a href="#6" style="background-color:brown; color: #ffffff; padding: 7px 10px; text-decoration: none; border-radius: 50px;">Back to topic</a>

<a id="6.2"></a>
### <b><span style='color:green'>6.2 Using Seaborn</span></b>

```python
sns.set(style="darkgrid")
sns.scatterplot(data=df, x="area_square_ft", y="price")
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='area_square_ft', y='price', scatter_kws={'s': 100, 'alpha': 0.9, 'color': 'orange'}, line_kws={'color': 'red'})

plt.title('Area vs. Price', fontsize=16)
plt.xlabel('Area (sq ft)', fontsize=12)
plt.ylabel('Price', fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()
```

In above code, I used 'sns.regplot()' to fit a regression line to the scatter plot.
