## <b><span style='color:red'>Outlier Detection and Removal</span></b> 

```python
mean = df3.height.mean()
std_deviation = df3.height.std()
mean, std_deviation
```

<a id="4.1"></a>
### <b><span style='color:red'>Outlier detection and removal using Standard Deviation</span></b>

Standard Deviation measures the dispersion of data by computing the distance of the data to its mean

```python
min_threshold, max_threshold = mean-3*std_deviation, mean+3*std_deviation
min_threshold, max_threshold
```

```python
# outliers
df3[(df3.height < min_threshold) | (df3.height > max_threshold)]
```

```python
df3_no_outlier = df3[(df3.height<77.91) & (df3.height>54.82)]
df3_no_outlier.shape
```

<a id="4.1"></a>
### <b><span style='color:red'>Outlier detection and removal using Z score</span></b>

Z score is another way to remove outliers.

Z score indicates how many standard deviation away a data point is from mean. For example in our case mean is 66.37 and standard deviation is 3.84. If a value of a data point is 77.91 then Z score for that is 3 because it is 3 standard deviation away (77.91 = 66.37 + 3 * 3.84)

We calculate the Z Score as follows:



```python
df3['zscore'] = ( df3.height - df3.height.mean() ) / df3.height.std()
df3.head(5)
```

```python
# one side outliers
df3[df3['zscore']<-3]
```

```python
# other side outliers
df3[df3['zscore']>3]
```

```python
df3_outlier_removed = df3[df3['zscore']>-3]
df3_outlier_removed = df3[df3['zscore']<3]
df3_outlier_removed.shape
```

## <b><span style='color:red'>Modified Z score</span></b>

```python
df4 = pd.read_csv("/kaggle/input/aiml-dataworld/movie_revenues.csv")
display(df4)
```

```python
df4.revenue.describe()
```

Here revenue is an absolute dollar value. To avoid a large scale, we will add a new column for **revenue in millions**.

```python
df4['revenue_mln'] = df4['revenue'].apply(lambda x: x/1000000)
df4.revenue_mln.describe()
```

```python
_, mean, std, *_ = df4.revenue_mln.describe()
mean, std
```

<a id="9.1"></a>
### <b><span style='color:red'>Outlier detection using Z score</span></b>

```python
def get_z_score(value, mean, std):
    return (value - mean)/std
```

```python
df4['z_score'] = df4.revenue_mln.apply(lambda x: get_z_score(x, mean, std))
df4.head()
```

General guideline is to treat anything that has a z score of 3 or more as an outlier

```python
df4[df4.z_score>3]
```

<a id="9.2"></a>
### <b><span style='color:red'>Outlier detection using modified Z score</span></b>

As we got only 1 outlier using simple Z score. Based on the kind of analysis we are doing sometimes it might be better to use modified Z score as it will return more outliers.

```python
def get_mad(s):
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df4.revenue_mln)
median = np.median(df4.revenue_mln)
MAD, median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2711, median, MAD)
```

```python
df4['mod_z_score'] = df4.revenue_mln.apply(lambda x: get_modified_z_score(x, median, MAD))
df4.head()
```

General guideline for modified Z score is to use 3.5 as a thresold. i.e. anything that has a mod z score of 3.5 or more is an outlier

```python
df4[df4.mod_z_score>3.5]
```
